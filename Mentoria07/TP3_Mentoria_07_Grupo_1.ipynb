### Metricas

1- ¿Qué métricas existen para evaluar modelos de regresión?



Existente numerosas métricas para evaluar los modelos de Regresión:
Algunas Métricas de Regresión:
* Error cuadrático medio (MSE): mide el error cuadrado promedio de nuestras predicciones.
* Raíz del Error cuadrático medio (RMSE):es solo la raíz cuadrada de MSE. La raíz cuadrada se introduce para hacer que la escala de los errores sea igual a la escala de la variable a predecir.
* Error absoluto medio (MAE): el error se calcula como un promedio de diferencias absolutas entre los valores objetivo y las predicciones.
* El coeficiente de determinación (R²): mide cuanto de la variabilidad de la respuesta se debe a la variabilidad del predictor.
* R cuadrado ajustado (R²(n-1))
* Indice de concordancia de Lin (CCC): el coeficiente de correlación de concordancia mide la concordancia entre dos variables, por ejemplo, para evaluar la reproducibilidad o la confiabilidad entre evaluadores.
* Error de porcentaje cuadrático medio (MSPE)
* Error porcentual absoluto medio (MAPE)
* Error logarítmico cuadrático medio (RMSLE)


2- ¿Cuál de esas métricas es más conveniente utilizar y porqué?

Nos parece conveniente usar:
* Raiz del error cuadratico medio ya que se encuentra en la misma magnitud que la variable target al tomar la raiz. Evaluo la exactitud.

* Coeficiente de determinación (R2): ya que evalua la variabilidad explicada por los predictores y por lo tanto indica la precisión de la estimación.

A continuación una breve descripción de los posibles modelos:

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

- **LGLM Regressor** (Lasso Generalized Linear Model Regressor):

El LGLM Regressor es un modelo de regresión basado en el algoritmo Lasso (Least Absolute Shrinkage and Selection Operator) aplicado a modelos lineales generalizados. Utiliza la regularización L1 para penalizar los coeficientes de las características menos importantes, lo que puede llevar a la selección automática de características.
Uso: Es útil cuando se desea un modelo de regresión lineal que tenga la capacidad de seleccionar automáticamente las características más relevantes y reducir el riesgo de sobreajuste.

- **Regresión lineal de ElasticNet**:

La regularización combina las penalizaciones L1 (Lasso) y L2 (Ridge). Donde ridge tiene el efecto de reducir de forma proporcional el valor de todos los coeficientes del modelo pero sin que estos lleguen a cero, mientras que Lasso fuerza a que los coeficientes de los predictores tiendan a cero, de aquellos predictores que no aportan a la explicación de la respuesta.
- **Decision Tree Regressor**:

Es un modelo basado en árboles de decisión, utilizado para resolver problemas donde el objetivo es predecir un valor numérico. Funciona dividiendo repetidamente el conjunto de datos en subconjuntos más pequeños basados en características específicas. A medida que se desciende por el árbol, se toman decisiones basadas en reglas que dividen los datos en función de las características que mejor separan los valores objetivo. La predicción se obtiene haciendo el promedio de la variable objetivo o target, en el conjunto de datos que cumplen con esas características. Este enfoque permite capturar relaciones no lineales en los datos, sin embargo, tienden a ser propensos al sobreajuste, lo que puede mitigarse mediante técnicas de poda o el uso de ensamblados como Random Forest.

- **Random Forest Regressor**:

El Random Forest Regressor se basa en el concepto de un "bosque" de árboles de decisión. Cada árbol en el bosque se entrena con una muestra aleatoria de los datos de entrenamiento y una selección aleatoria de características. Luego, los resultados de todos los árboles se promedian para predecir el valor de regresión final.
Uso: Ayuda a reducir el sobreajuste y a mejorar la generalización del modelo. El Random Forest es conocido por su robustez y capacidad para manejar conjuntos de datos grandes con muchas características.

- **Gradient Boosting Regressor**:

El Gradient Boosting Regressor es un algoritmo de aprendizaje automático que combina múltiples modelos de regresión débiles para formar un modelo de regresión fuerte. Utiliza el enfoque de boosting para mejorar el rendimiento del modelo, minimizando iterativamente los errores de predicción en cada etapa.
Uso: Es útil en una amplia variedad de tareas de regresión y a menudo se considera uno de los algoritmos más potentes para este propósito. Puede manejar datos complejos y no lineales de manera efectiva.

- **XGboost Regression**(extreme gradient boosting):

Los conjuntos se construyen a partir de modelos de árboles de decisión. Los árboles se agregan uno por uno al conjunto y se ajustan para corregir los errores de predicción cometidos por modelos anteriores. Este es un tipo de modelo de aprendizaje automático conjunto denominado impulso.
Los modelos se ajustan utilizando cualquier función de pérdida diferenciable arbitraria y algoritmo de optimización de descenso de gradiente. Esto le da a la técnica su nombre, " impulso de gradiente ", ya que el gradiente de pérdida se minimiza a medida que se ajusta el modelo, de manera muy similar a una red neuronal.

A continuación listamos los que consideramos más importante y que vamos a ajustar:


**LGLM Regresor**:
- *alpha*: es la constante que multiplica el término L1, controlando la fuerza de regularización. Puede valer entre 0 e infinito, el defoult es 1.
- *max_iter*: el número máximo de iteraciones.
- *tol*: es la tolerancia para la convergencia del algoritmo. El modelo se considera convergido cuando la mejora en la función de pérdida es menor que este valor. El valor por defoult es 1e-4.

**Elastic Net Regresor**:

- *alpha*: Constante que multiplica los términos de penalización. Por defecto es 1.0.
- *l1_ratio*: El parámetro de mezcla de la regularización,toma valores entre 0 y 1. Cuando l1_ratio = 0 la penalización a L2  es es la penalidad lasso. Para l1_ratio = 1 es una penalización componente a componente L1.Para 0 < l1_ratio < 1, la penalización es una combinación de L1 y L2.
- *max_iter*: el número máximo de iteraciones.
- *tol*: es la tolerancia para la convergencia del algoritmo. El modelo se considera convergido cuando la mejora en la función de pérdida es menor que este valor. El valor por defoult es 1e-4.

**Decision Tree Regressor**:

- *criterion*:  es la función para medir la calidad de una división. Puede ser "squared_error" (usa la media, es más sensible a valores atípicos); "absolute_error” (usa la mediana);  “friedman_mse” (utiliza el error cuadrático medio con la puntuación de mejora de Friedman para posibles divisiones), “poisson” (utiliza la reducción de la desviación de Poisson para encontrar divisiones).
- *max_depth*: máxima profundidad del arbol.
- *min_samples_split*: cantidad mínima de muestras necesarias para que se vuelva a dividir un nodo.
- *max_features*: número máximo de variables que se analizan en cada división de nodo.
- *min_samples_leaf*: Es el número mínimo de muestras necesarias para estar en un nodo de hoja. Solo se considerará un punto de división (a cualquier profundidad) si deja al menos min_samples_leaf muestras  en cada una de las ramas (izquierda y derecha).
- *max_features*: cant. max. de variables que considera en cada división.

**Random Forest Regressor**:

- *n_estimators*: número de árboles que se utilizará. El valor por defoult es 100.
- *max_depth:*
- *min_samples_split*
- *min_samples_leaf*
- *max_features*

**Gradient Boosting Regressor**:

- *learning_rate*: controla la tasa de aprendizaje del modelo. Determina cuánto se ajustarán los pesos de los estimadores (árboles) en cada iteración. Valor por defoult: 0,1.
- *n_estimators*
- *max_depth:*
- *min_samples_split*
- *min_samples_leaf*
- *max_features*

**XGboost Regression**:

- *learning_rate*: (también conocido como eta): Controla la tasa de aprendizaje y ajusta la contribución de cada árbol al modelo final. Un valor más bajo hace que el modelo sea más robusto pero requiere más árboles para alcanzar un buen rendimiento.
- *n_estimators*
- *max_depth*
-*min_child_weight*: Controla la cantidad mínima de observaciones necesarias en un nodo hoja. Puede ayudar a evitar la partición excesiva y el sobreajuste.
-*gamma (también conocido como min_split_loss)*: Un valor mínimo requerido para realizar una partición en un nodo hoja. Puede ayudar a evitar particiones que no mejoran significativamente la pérdida.

```python
# Definimos los modelos y sus respectivos hiperparámetros:
models = {
    'Lasso': {
        'model': Lasso(),
        'params': {
            'alpha': [0.001, 0.003, 0.01]
        }
    },
    'ElasticNet': {
        'model': ElasticNet(),
        'params': {
            'alpha': [0.001, 0.003, 0.01],
            'l1_ratio': [0.3, 0.5, 1.0]
        }
    },
    'Decision Tree': {
        'model': DecisionTreeRegressor(),
        'params': {
            'max_depth': [8, 15, 30],
            'min_samples_split': [20, 50, 80],
            'min_samples_leaf': [1, 5, 10]
        }
    },
        'RandomForest': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [50, 100, 200],
            'max_depth': [8, 15, 30],
            'min_samples_split': [20, 50, 80],
            'min_samples_leaf': [1, 5, 10],
            'random_state': [42]
        }
    },
        'GradientBoostingRegressor': {
        'model': GradientBoostingRegressor(),
        'params': {
            'n_estimators': [50, 100, 200],
            'max_depth': [8, 15, 30],
            'min_samples_split': [20, 50, 100],
            'min_samples_leaf': [1, 5, 10],
            'learning_rate': [0.05, 0.1, 0.2],
            'random_state': [42]
        }
    },
        'XGBoost': {
        'model': XGBRegressor(),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.001],
            'max_depth': [5, 10, 15],
            'min_child_weight': [1, 2, 3],
            'gamma': [0, 0.1, 0.2],
        }
    }
}
```

### Test de modelos

Cabe aclarar que no han sido corridos en el mismo notebook por cuestiones de división del costo computacional pero todos los data set pasaron por el mismo ciclo for

```python
results_m1_s3 = {}  # Para almacenar los resultados

for model_name, model_config in models.items():
    # Registra el tiempo de inicio
    start_time = time.time()

    # Define los hiperparámetros que deseas ajustar
    param_grid = model_config['params']

    # Configura RandomizedSearchCV para el modelo y los hiperparámetros
    random_search = RandomizedSearchCV(model_config['model'], param_grid,
                                     n_iter=15, scoring='r2', cv=5,
                                     random_state=42)

    # Realiza la búsqueda de hiperparámetros en los datos de entrenamiento
    random_search.fit(X_train_m1_s3, y_train_m1a4)

    # Registra el tiempo de finalización
    end_time = time.time()

    # Calcula el tiempo transcurrido en segundos
    elapsed_time = end_time - start_time

    # Predecir en el conjunto de prueba
    best_estimator = random_search.best_estimator_
    y_pred_val = best_estimator.predict(X_val_m1_s3)
    y_pred_train = best_estimator.predict(X_train_m1_s3)

   # Calcular la precisión del modelo
    mse_val = mean_squared_error(y_val_m1a4, y_pred_val)
    mse_train = mean_squared_error(y_train_m1a4, y_pred_train)
    r2_val = r2_score(y_val_m1a4, y_pred_val)
    print(f"mse: {mse_val:.2f} , r2: {r2_val:.2f} ")
    # Almacena los resultados

    results_m1_s3[(model_name)] = {
        'best_params': random_search.best_params_,
        'r2_train': random_search.best_score_,
        'r2_val': r2_val,
        'mse_train': mse_train,
        'mse_val': mse_val,
        'elapse_time':  elapsed_time
    }

results_m1_s3_df = pd.DataFrame(results_m1_s3)
results_m1_s3_df.to_csv('results_m1_s3.csv', index=False)
files.download('results_m1_s3.csv')
```

Esto es razonable ya que a mayor dimensión mayor gasto, sin embargo quí resulta de interés observar como la variable target es de *consumo energético por unidad de área*, se observa que el tamaño del edificio (es decir la superficia total construida) es una variable muy importante en la prediciñon, indicando que no es lineal el consumo por unidad de area (si fuese asi todos los edificios tendrian el mismo consumo por unidad de area). Esto es muy im portante a la hora de tomar decisiones sobre los tamaños a contruir.

En segunda instancia aparece el ranking de eficiencia. Si bien se esperaba que esta variable fuera muy explicativa, el desafío fue la cantidad de valores faltantes de esta variable. El modelo fue exitoso en imputar datos faltantes y capturar poder de predicción.
En este sentido, otra variable con datos faltantes (aunque mucho menor porcentaje) es el año de edificación que aparece como 3° variable de importancia. Esto puede deberse a las tecnologías y materiales utilizados que al avanzar en el tiempo deberían estar más adecuados a prestaciones de menor consumo.

Luego tenemos un grupo de variables referidas al tipo de edificio, entre las cuales aparecen: centros de salud, otros, depósito y residencial. Algunas conjeturas que podemos hacer son: Los centros de salud tienen seguramente grandes consumo debido al tipo de actividad. En “otros” están incluidos grupos como data center, con grandes consumos (aunque en dc0 ese grupo estaba excluido). Los depósitos por otro lado seguramente tengan grandes superficies y menos consumo.

Entre las últimas variables aparece la altura sobre el nivel del mar, una variable de la cual no podemos tener explicaciones tan intuitivas como las demás, pero podríamos especular que a mayor altura, se espera que las temperaturas sean más bajas y  la temperatura media de invierno se encuentra entre las variables explicativas. Esto también puede ser importante a la hora de decidir dónde realizar o no una construcción.

Por último, podemos decir respecto a las variables climáticas que las mismas aportan muy poca o nula explicación, excepto por la de temperatura medias de la época mas fría, indicando que el gasto de energía es mayor para calefaccionar que para enfriar en épocas cálidad. Esto está en corcondacia con la observación que hicimos en el 1 TP de que las variables referidas al frio tienen mucha mayor correlación con la variable target que las referidas al calor, lo cual estaría indicando que el mayor consumo se debe al hecho de calefaccionar y no tanto de enfriar. Esto también puede ser importante a la hora de diseñar edificios, que tengan en cuenta la variable frio.

En conclusión, respecto a cómo se configura el modelo y cuáles son las variables más explicativas y, por lo tanto, aquellas que tendremos que tener bien definidas para hacer buenas predicciones al recibir nuevos datos son los siguientes: las dimensiones de edificio mismo, como rankea en eficiencia, su año de construcción; la función del edificio, la elevacion sobre el nivel del mar y las temperaturas medias de la estaciones más frías.
